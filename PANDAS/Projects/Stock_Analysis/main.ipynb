```python
import yfinance as yf

# Define the stock symbol and time period
ticker = 'AAPL'  # Apple Inc.
stock = yf.Ticker(ticker)

# Fetch historical data
df = stock.history(period='1y')  # Last 1 year
print(df.head())
```

```python
import pandas as pd

print("Summary stats:")
print(df.describe())

print("\nMissing values:")
print(df.isnull().sum())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 6))
plt.plot(df['Close'], label='Close Price')
plt.title('Apple Stock Closing Prices - Last 1 Year')
plt.xlabel('Date')
plt.ylabel('Price ($)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df['MA20'] = df['Close'].rolling(window=20).mean()
df['MA50'] = df['Close'].rolling(window=50).mean()

plt.figure(figsize=(14, 6))
plt.plot(df['Close'], label='Close Price')
plt.plot(df['MA20'], label='20-Day MA')
plt.plot(df['MA50'], label='50-Day MA')
plt.title('Apple Stock Price with Moving Averages')
plt.legend()
plt.show()
```

```python
df['Daily Return'] = df['Close'].pct_change()

plt.figure(figsize=(14, 6))
plt.plot(df['Daily Return'], label='Daily Return')
plt.title('Daily Returns of Apple Stock')
plt.legend()
plt.show()
```

```python
symbols = ['AAPL', 'MSFT', 'GOOGL']
data = yf.download(symbols, period='1y')['Close']

data.plot(figsize=(14, 6))
plt.title("Stock Prices - AAPL vs MSFT vs GOOGL")
plt.ylabel("Price ($)")
plt.grid(True)
plt.show()
```
